```python
#make variable for features with missing
features_with_na = [features for features in df.columns if df[features].isnull().sum()>0]
features_with_na
```

```python
# I want to make a little df to visualize missing values
missing_overview = pd.DataFrame({'features_with_na':features_with_na})
missing_overview['count'] = [df[feature].isnull().sum() for feature in features_with_na]
missing_overview['percentage'] = [np.round(df[feature].isnull().mean()*100,3) for feature in features_with_na]
missing_overview
```

Now I want to actully get a full sense of the features before fixing them and filling the nulls, so first:
### Full EDA

```python
 # See distribution of SalesPrice
sns.histplot(x = df['SalePrice'], bins =25)
plt.title('Sale Price', fontdict = {'fontsize': 15})
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.show()

#Data seems to be right skewed
#This means that the mean has been pulled to the right of median and mode by the tail on the right
#because of this I believe I should use the median when calling for Sale Price as it is a better metric of center
```

```python
#A lot of the rest of the EDA from here on was inspired by the thought process and code of a DataScientist I follow Krish Naik
#These are mostly things I was already going to do but maybe the order/organization, and some 
#of the methods of visualization of how he approached it connected with me
#Video link https://www.youtube.com/watch?v=ioN1jcWxbv8&t=867s

#I want to understand the relationship between features with missing values and the sales price 
for feature in features_with_na:
    local_df = df.copy()
    
    # to make a variable that indicates 1 if the observation was missing or zero otherwise
    local_df[feature] = np.where(local_df[feature].isnull(), 1, 0)
    
    #now groups by whether there was a null or not and pulls the Sales price median
    #I initially used mean but median makes more sense in context so that houses with price outliers dont pull 
    feature_series = local_df.groupby(feature)['SalePrice'].median()
    
    # let's calculate the mean SalePrice where the information is missing or present
    plots = sns.barplot(x = feature_series.index,y= feature_series.values, palette = 'Greens');
    
    #This next bit adds the annotations
    for bar in plots.patches:
        plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()*0.85), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
    plt.title(feature)
    plt.show()
    
    #Help annotating from https://www.geeksforgeeks.org/how-to-annotate-bars-in-barplot-with-matplotlib-in-python/
    
# Each chart compares the median sale price for entries with no null values(0) and ones that did have a null/missing value(1)
```

```python
#Seems to be a clear visible relationship with almost all of these features and SalePrice with whether they were missing or not
```

```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O' and feature not in ['Id']]
print(len(numerical_features))
numerical_features
```

```python
non_numerical_features = [feature for feature in df.columns if df[feature].dtypes == 'O' ]
print(len(numerical_features))
numerical_features
```

```python
# Outline important DateTime features
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
#Visualize relation to Yr Sale
df.groupby('Yr Sold')['SalePrice'].median().plot.line();
```

```python
 ## graph each other date features in years relative to the year sold

for feature in year_features:
    if feature!='Yr Sold': 
        local_df=df.copy()
       
        local_df[feature]=local_df['Yr Sold']-local_df[feature] #Grabs this difference from relevant year and year sold

        plt.scatter(local_df[feature],local_df['SalePrice'])
        plt.xlim(left = -5)# Garage had a wild negative outlier
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Now to explore discrete and continuous features 
discrete_features =[feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_features+['Id']]

print(len(discrete_features))
discrete_features
```

```python
for feature in discrete_features:
    grouping = df.groupby(feature)['SalePrice'].median()
    sns.barplot(x = grouping.index, y = grouping.values, palette = 'rocket')
    plt.title(feature)
    plt.show()
```

```python
#Some really cool trends already seen in the features in respect to Sale price
```

```python
continuous_features = [feature for feature in numerical_features if len(df[feature].unique())>25 and feature not in year_features+['Id']]

print(len(continuous_features))
continuous_features
```
